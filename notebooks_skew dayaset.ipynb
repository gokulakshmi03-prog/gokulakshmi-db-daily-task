```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("e-waste_data_2022_complete.csv")
print(df)
```

```python
skew_value=df["Recycling Rate (%)"].skew()
print("\nskewness value :",round(skew_value,4))
```

```python
if skew_value > 0:
    print("positive skew distribution")
elif skew_value <0:
    print("negative skew distribution")
else:
    print("symmetrical distribution")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Recycling Rate (%)"],bins=7)
plt.xlabel("profit")
plt.ylabel("frequency")
plt.title("e waste management")
plt.grid(True)
plt.show()
```
